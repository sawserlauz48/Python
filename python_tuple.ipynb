In Python, a tuple is a built-in data structure that is similar to a list in many ways but has a key difference: tuples are immutable. This means that once you create a tuple, you cannot change its contents - you cannot add, remove, or modify elements within a tuple. Tuples are typically defined by enclosing a comma-separated sequence of values within parentheses ().

```python
# Accessing Elements
colors=("red","green","blue")
print(colors[2])
```

```python
# Slicing
colors=("red","green","blue")
print(colors[0:2])
```

```python
# Iteration
colors=("red","green","blue")
for color in colors:
  print(color)
```

```python
# Concatenation
colors=("red","green","blue")
colors2=("cyan", "magenta", "yellow")
print(colors+colors2)
```

```python
# Membership
colors=("red","green","blue")
if "blue" in colors:
  print("blue is in the tuple")
```

```python
# Finding Length
colors=("red","green","blue")
print(len(colors))
```

```python
# Immutability
colors=("red","green","blue")
colors[2]="yellow"
print(colors)
 ## this will return error
 ## you cant change elements inside a tuple after creation
```

```python
# Hashability
colors={("primary_colors"):"RGB",("secondary_color"):"CMY"}
print(colors)

```
